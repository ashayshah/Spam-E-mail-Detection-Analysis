```python
# Ignore  the warnings
import warnings
warnings.filterwarnings('ignore')


#Question 5a - fetching data
import pandas as pd
dt = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data', header=None)
#display(dt)




# Question 5b and c
# Converting header file .name to a dataframe and merging it with the .data file. This will create a new dataframe with headers and data together.
data1 = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.data', header=None)
#'data1' stores the data minus the headers
#display(data1)

data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/spambase/spambase.names',error_bad_lines=False)
#'data' stores the file with headers
#display(data)

x = data.loc[18:] #slicing is done. From 18th row, to the last row.. Because only those rows contain headers.
x['| SPAM E-MAIL DATABASE ATTRIBUTES (in .names format)'] = x['| SPAM E-MAIL DATABASE ATTRIBUTES (in .names format)'].str.replace("continuous.", "")
#removing the word 'continuous'.
#display(x)

col_one_list = x['| SPAM E-MAIL DATABASE ATTRIBUTES (in .names format)'].tolist()
col_one_list.append('spam/no_spam') #appending this new header on the 57th line
#display(col_one_list)

df = pd.DataFrame(col_one_list)
#display(df)
df_t = df.transpose()
#display(df_t)

frames = [df_t,data1]
#display(frames)
result = pd.concat(frames) #concatenating header file and data file, so basically we merge it and make it into 1 dataframe
#display(result)

#to make the 1st row of words as our main column headers :
new_header = result.iloc[0] #grab the first row for the header
#print(new_header)
result = result[1:] #take the data minus the new_header row
#display(result)
result.columns = new_header
#display(result.columns)

#display(result)
#so we took the actual headers in the 1st row, made a variable new_header and stored our headers in it.##
##appended the new variable to the result variable as our new column

x = result.columns.tolist()
#display(x)

#converting all decimal entries more than 0, directly into 1
for i in x:
    result.loc[result[i] > 0, i] = 1
display(result)



#Question 5d,f
#Importing sci-kit learn libraries

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB,BernoulliNB
%matplotlib inline

#splitting the data into 80% as train set
spam = result[result['spam/no_spam'] == 1]
not_spam = result[result['spam/no_spam'] == 0]
spam_train, spam_test = train_test_split(spam, train_size=0.8)
not_spam_train, not_spam_test = train_test_split(not_spam, train_size=0.8)
X_train = not_spam_train.append(spam_train)
X_train = X_train.astype('int')
y_train = X_train.pop('spam/no_spam')
y_train=y_train.astype('int')
X_test = not_spam_test.append(spam_test)
X_test = X_test.astype('int')
y_test = X_test.pop('spam/no_spam')
y_test = y_test.astype('int')

#without laplace smoothing
spam_bayes = BernoulliNB(alpha=0.0, class_prior=None, fit_prior=True)
spam_bayes.fit(X_train, y_train)
spam_bayes.score(X_train, y_train)
spam_bayes.score(X_train, y_train)

#With laplace smoothing
spam_bayes = BernoulliNB(alpha=1.0, class_prior=None, fit_prior=True)
spam_bayes.fit(X_test, y_test)

predict = spam_bayes.predict(X_test)



from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

Cf = confusion_matrix(y_test,predict)
print(Cf) #true positive, true negative, false positive & false negative.

print(classification_report(y_test, predict,))

accuracy = sum(Cf.diagonal())/(Cf[0,0]+Cf[0,1]+Cf[1,0]+Cf[1,1])
# we printed the accuraxy, which is : (True positive + true negative)/ Tp + Tn + Fp + Fn
print(accuracy)
spam_bayes.score(X_test, y_test)

spam_bayes.score(X_train, y_train)
```
